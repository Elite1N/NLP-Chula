```python
# Chech data leakage
import pandas as pd
dev_split = pd.read_csv("../data/dev_split.csv")
train_split = pd.read_csv("../data/train_split.csv")
train_split_enriched = pd.read_csv("../data/train_split_enriched.csv")

# Find exact row matches across all shared columns
common_cols = [c for c in dev_split.columns if c in train_split.columns]

matched_rows = dev_split.merge(train_split, on=common_cols, how="inner")

print(f"Shared columns used for match: {len(common_cols)}")
print(f"Number of matching rows: {len(matched_rows)}")

if not matched_rows.empty:
    print("\nSample matching rows:")
    print(matched_rows.head())
else:
    print("No matching rows found.")
```
