```python
import pandas as pd
DATA_PATH = "../data/"
df = pd.read_csv(DATA_PATH+"dev_set.csv")

df.head()
```

```python
from pathlib import Path

# Locate train.tok next to the current dev file
train_tok_path = Path(DATA_PATH + "train.src.tok")
print("=== DEV SET (from df) ===")
print(f"path: {DATA_PATH}")
print(f"shape: {df.shape} (rows, cols)")
print("\ncolumns:")
print(df.dtypes)
print("\nmissing values per column:")
print(df.isna().sum())
print("\npreview:")
display(df.head())

print("\n=== TRAIN.TOK ===")
print(f"path: {train_tok_path}")

if train_tok_path.exists():
    size_bytes = train_tok_path.stat().st_size
    with train_tok_path.open("r", encoding="utf-8", errors="replace") as f:
        lines = f.readlines()

    print(f"size (bytes): {size_bytes}")
    print(f"line count: {len(lines)}")
    print("\nfirst 5 lines:")
    for i, line in enumerate(lines[:5], 1):
        print(f"{i}: {line.rstrip()}")
else:
    print("train.tok not found at the expected path.")
```
